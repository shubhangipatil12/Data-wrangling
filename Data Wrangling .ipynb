```python
#Combining all csv files into one using glob and enumerate
import glob, os
import pandas as pd
 
os.chdir("C:/Users/shubhangipatil/Desktop/Python")
results = pd.DataFrame([])
 
for counter, file in enumerate(glob.glob("articles*")):
    namedf = pd.read_csv(file, skiprows=0)
    results = results.append(namedf)
results.to_csv('C:/Users/shubhangipatil/Desktop/Python/combinedarticles1.csv')
```

```python
# Data wrangling steps with Articles dataset

#First step: Checking for missing, null values

#Using index_col=0 to make id as the first column instead of title as first column

df=pd.read_csv('C:/Users/shubhangipatil/Desktop/Python/combinedarticles1.csv',index_col=0)
pd.isnull(df).any()
```

```python
#Dropping columns like Unnamed and Id which are not useful for analysis

df.drop(['id'],axis=1)


#Drop column with null values

df2=df.drop(['url'],axis=1)

#Drop rows which have author as missing values

df3=df2.dropna(subset=['author'])

# df4=df3.isnull()

# pd.isnull(df4).any()
```

```python
#Filtering data based on conditions

#Getting articles based on specific months

df3[df3['month']==12]
```

```python
#Getting data as per the condition shown below
df3[df3['publication']=='Vox']
```

```python
#Grouping data based on specific columns or values

df3.groupby('publication').author.value_counts()
```

```python
#Alternatives to missing values in columns containing numbers

# Fill all missing values with the mean of the particular column
df.fillna(df.mean())

# Fill any missing value in column 'A' with the column median
df['A'].fillna(df['A'].median())

# Fill any missing value in column 'B' with the column mode
df['B'].fillna(df['B'].mode())
```
